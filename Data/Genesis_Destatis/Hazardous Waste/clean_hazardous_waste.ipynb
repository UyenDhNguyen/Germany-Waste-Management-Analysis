## Hazardous Waste-producers-quantities-years-industry-types of waste(EAV 2).csv
Table 32151-0001
Columns: Year, Industry Code, Industry, EAV code, Waste Description, Number of waste generators, Volumes of waste (1000t), Volumes of waste generated by primary producers (1000t)
Read Lines: 9 - 10298
To Translate: Industry, Waste Description

```python
columns_names = "Year, Industry Code, Industry, EAV code, Waste Description, Number Of Waste Generators, Volumes Of Waste (1000t), Volumes Of Waste Generated By Primary Producers (1000t)"
columns_names = columns_names.split(", ")
columns_names
```

```python
filename4 = "Hazardous Waste-producers-quantities-years-industry-types of waste(EAV 2).csv"
waste_df4 = pd.read_csv(filename4, sep=";", header=None, encoding="latin-1", names=columns_names, skiprows=8, nrows=10298-8, na_values="-")
waste_df4.tail()
```

```python
waste_df4.head()
```

```python
# Translate Industry
industries = waste_df4['Industry'].unique()
industries_en = (translator.translate(x).text for x in industries)
industries_translation = dict(zip(tuple(industries), industries_en))
waste_df4['Industry'] = waste_df4['Industry'].apply(lambda x: industries_translation[x])
waste_df4['Industry'] = waste_df4['Industry'].str.title()
waste_df4
```

```python
waste_df4.info()
```

```python
waste_df4['Volumes Of Waste (1000t)'] = waste_df4['Volumes Of Waste (1000t)'].apply(lambda x: to_float(x))
waste_df4['Volumes Of Waste Generated By Primary Producers (1000t)'] = waste_df4['Volumes Of Waste Generated By Primary Producers (1000t)'].apply(lambda x: to_float(x))
```

```python
waste_df4['Waste Description'] = waste_df4['Waste Description'].apply(lambda x: translator.translate(x).text)
waste_df4.to_csv("cleaned_32151-0001_producers-quantities-years-industry-types of waste(EAV 2).csv", index=False)
```
