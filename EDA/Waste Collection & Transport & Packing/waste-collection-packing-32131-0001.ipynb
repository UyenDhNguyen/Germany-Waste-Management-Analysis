```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
filename = "../../Data/Genesis_Destatis/Waste Collection & Transport & Packing/cleaned_32131-0001_Waste-Collection-Transport-Packing-waste collection-transport-packing.csv"
df = pd.read_csv(filename)
print(df.shape)
df.head()
```

```python
df['Year'] = df['Year'].astype('str').astype('category')
```

## Plot KPI through years

```python
def plot_kpi_years(df, kpi):
    plt.figure(figsize=(15, 7))
    sns.lineplot(df, x='Year', y=kpi)
    plt.title(f"{kpi} Through Years")
    if kpi == "Collected Transport And Secondary Packaging (1000t)":
        plt.ylabel("Amount (1000t)")
    else:
        plt.ylabel("Count")

plot_kpi_years(df, "Collected Transport And Secondary Packaging (1000t)")
```

Collected Transport And Secondary Packaging varied wildly through years, with the steepest slope from 2008 to 2012 (probably it had sth to do with the recession), and rose above 4700000 tonnes in the recent years from 2015 to 2020

```python
plot_kpi_years(df, "Number Of Collectors")
```

The number of collectors decreased steadily from 2002 to 2017, followed by a strange spike in 2018 before dropping again in the following years

```python
plt.figure(figsize=(7, 7))
sns.scatterplot(df, x='Number Of Collectors', y='Collected Transport And Secondary Packaging (1000t)')
plt.title("Number Of Collectors Against Collected Transport And Packaging")
```
